```python
import pandas as pd
import json
import matplotlib.pyplot as plt
import os

# Caminho para o CSV dentro de outputs
csv_path = os.path.join('outputs', 'experiments_log.csv')

if os.path.exists(csv_path):
    df = pd.read_csv(csv_path)
    
    # Pegar o melhor modelo
    best_row = df.sort_values(by='best_valid_acc', ascending=False).iloc[0]
    print(f"Melhor Acc Teste: {best_row['final_test_accuracy']:.4f}")
    
    # Abrir o hist√≥rico JSON associado
    with open(best_row['history_file'], 'r') as f:
        hist = json.load(f)
        
    # Plotar
    plt.plot(hist['valid_acc'], label='Val Acc')
    plt.plot(hist['test_acc'], label='Test Acc')
    plt.legend()
    plt.show()
else:
    print("Nenhum log encontrado ainda.")
```
